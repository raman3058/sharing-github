```python
#Import Pandas
import pandas as pd

#Load Datasets
cc_apps = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data", header=None)

#Inspect data to print first 5 rows.
cc_apps.head()
```

```python
#Adding columns name to the dataframe of cc_apps:

cc_apps.columns = ['Gender', 'Age', 'Debt', 'Married', 'BankCustomer', 'EducationLevel', 'Ethincity', 'YearsEmployed', 'PriorDefault', 'Employed', 'CreditScore', 'DriversLicense', 'Citizen', 'ZipCode', 'Income', 'ApprovalStatus']
```

```python
cc_apps.head()
```

```python
#print summary statistics

cc_apps_description = cc_apps.describe()
print(cc_apps_description)

print("\n")

#print dataframe information
cc_apps_info = cc_apps.info()
print(cc_apps_info)

print("\n")

#Inspect missing values in the dataset

print(cc_apps.tail(17))
```

```python
#Import Numpy
import numpy as np

#Inspect missing value in the data set
print(cc_apps.tail(17))

print("\n")

#Replace the '?'s with NaN
cc_apps.replace('?', np.nan, inplace = True)

#Inspect the missing value again
print(cc_apps.tail(17))
```

```python
#Impute the missing values with mean imputation

cc_apps.fillna(cc_apps.mean(), inplace=True)

#To avoid the future warning & TypeError on calling the reduction, we can use the below one:-
#cc_apps.fillna(cc_apps[['Debt', 'YearsEmployed', 'CreditScore', 'Income']].mean(), inplace=True)

#Count the no of NaN's in datasets to verify

print(cc_apps.isna().sum())
```

```python
#Handling missing values for object type columns with most frequent values:-

for col in cc_apps.columns:
    if cc_apps[col].dtypes == "object":
        cc_apps = cc_apps.fillna(cc_apps[col].value_counts().index[0])
        
#Count the no of NaN's in datasets and print the count to verify.
print(cc_apps.isna().sum())
    
```

```python
#Convert non-numeric data to numeric
#import LabelEncoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

#Iterate over all the values of each columns & extract their dtypes

for col in cc_apps.columns.values:
    if cc_apps[col].dtypes == 'object':
        cc_apps[col] = le.fit_transform(cc_apps[col])
        
#printing the datatypes for verify
print(cc_apps.info())
```

```python
#Spliting datasets into train and test datasets
from sklearn.model_selection import train_test_split

#Drop the feature 11 and 13 as these two features are not required to include and convert the dataframe into numpy array 

cc_apps = cc_apps.drop(['DriversLicense', 'ZipCode'], axis=1).values
```

```python
#Segregate features and labels into separate variable
X= cc_apps[:, 0:13]
y = cc_apps[:, 13:14]

#reshape of y to flattened 1d array
y = y.ravel()

#Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#scaling X_train & X_test data before fit the model

from sklearn.preprocessing import MinMaxScaler

#Instantiate MinMaxScaler and rescale X_train & X_test

scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.fit_transform(X_test)
```

```python
#import Logistic Regression
from sklearn.linear_model import LogisticRegression

##Instantiate Logistic Regression classifier with default parameter values
logreg = LogisticRegression()

#Fit logreg to train the dataset
logreg.fit(rescaledX_train, y_train)

```

```python
#Import confusion martix

from sklearn.metrics import confusion_matrix

#use logreg to predict the instances from the test set and store it
y_pred = logreg.predict(rescaledX_test)

#accuracy score of logreg model
print("Accuracy of logistic regression classifier: ", logreg.score(rescaledX_test, y_test))

#print the confusion matrix of logreg model

print("Confusion matrix of logreg model:", "\n", confusion_matrix(y_pred, y_test))
```
